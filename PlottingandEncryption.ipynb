#Plots the sinusoidal functions with a legend and axes labels

```python
from matplotlib.pyplot import *

def plot_sinusoidal():
    """Plot sinusoidal functions and return a figure object."""
    x1 = numpy.linspace(0, 2 * math.pi, 75)
    x2 = numpy.linspace(0, 2 * math.pi, 75)
    f1 = numpy.sin(x1)
    f2 = numpy.cos(x1)
    
    figure()
    plot(x1, f1, 'o', label = 'graph of sinx')
    plot(x2, f2, 'o', label = 'graph of cosx')
    legend(loc = "upper left")
    
    return figure()

plot_sinusoidal()
```
